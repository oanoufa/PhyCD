```python
df_path = "/nfs/research/goldman/anoufa/figs/final_df_example.tsv"

# Read 20 samples from the file
df = pd.read_csv(df_path, sep="\t")

print(df.columns)
# Round float values
df['consensus'] = df['consensus'].round(2)
df['masked'] = df['masked'].round(2)
df['random'] = df['random'].round(2)
df['prop_gen_masked'] = df['prop_gen_masked'].round(2)
df['prop_dist_reduced'] = df['prop_dist_reduced'].round(2)
df['consensus_support'] = df['consensus_support'].round(2)
df['masked_support'] = df['masked_support'].round(2)

df.drop(columns=['prop_gen_masked', 'prop_dist_reduced'], inplace=True)

df_1 = df[['sample_name', 'consensus', 'masked', 'random', 'n_masked_cons',
       'n_masked_masked',
       'consensus_mutations_to_tree', 'masked_mutations_to_tree']]

df_2 = df[['consensus_placement', 'consensus_support', 'masked_placement',
       'masked_support', 'masked_mutations_masked']]

fig = df2img.plot_dataframe(
    df_1,
    col_width = [0, 2, 1, 1, 1, 2, 2, 4, 3, 3, 2, 3, 2, 8],
    tbl_header={
        "fill_color": "#D0DEBB",
        "font_color": "black",
        "line_width": 2,
        "line_color": "#D0DEBB",
        "font_size": 16,
    },
    tbl_cells={
        "font_color": "black",
        "font_size": 16,
        "line_width": 2,
        "line_color": "#000000",
        "height": 35,
    },
    row_fill_color=("#ffffff", "#D0DEBB"),
    fig_size=(1600, 675),
)

df2img.save_dataframe(fig=fig, filename="/nfs/research/goldman/anoufa/figs/final_df.png")

```
